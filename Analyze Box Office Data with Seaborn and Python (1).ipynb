### Task 2: Visualizing the Target Distribution

```python
fig, ax = plt.subplots(figsize=(16, 6)) #size of each plot
plt.subplot(1, 2, 1)   #one plot , 2 of them side by side(2 columns) , 1 row    confusion!
sns.distplot(train['revenue'], kde=False);
plt.title('distribution of revenue');
plt.subplot(1, 2, 2)
sns.distplot(np.log1p(train['revenue']), kde=False); #log1p numpy helping function , 0 maybe present and we dont want nan,infinity.
plt.title('distribution of log-transformed revenue');
```

```python
 train['log_revenue']= np.log1p(train['revenue'])  #new feature in train dataset  , log transformation of original distribution
```

### Task 3: Relationship between Film Revenue and Budget

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.scatterplot(train['budget'], train['revenue'])
plt.title('Revenue vs Budget')
plt.subplot(1, 2, 2)
sns.scatterplot(train['log_revenue'], np.log1p(train['budget']))
plt.title('log_Revenue vs log_Budget');
```

```python
train['budget'] = np.log1p(train['budget'])
test['budget'] = np.log1p(test['budget'])
```

### Task 4: Does having an Official Homepage Affect Revenue?

```python
train['homepage'].value_counts().head(10)
```

```python
train['has_homepage'] = 0 #inital by 0 means particular movie doesnot have homepage
train.loc[train['homepage'].isnull() == False, 'has_homepage'] = 1   #if particular movie has homepage then set the value to 1
test['has_homepage'] = 0
test.loc[test['homepage'].isnull() == False, 'has_homepage'] = 1
```

```python
sns.catplot(x='has_homepage', y='revenue', data=train);
plt.title('revenue of movies with and without homepage');
```

### Task 5: Distribution of Languages in Film

```python
language_data =train.loc[train['original_language'].isin(train['original_language'].value_counts().head(10).index)]
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.boxplot(x='original_language', y='revenue', data= language_data)
plt.title('mean revenue per language')
plt.subplot(1, 2, 2)
sns.boxplot(x='original_language', y='log_revenue', data= language_data)
plt.title('mean log revenue per language');
```

### Task 7: Do Film Descriptions Impact Revenue?

```python
import eli5
from sklearn.linear_model import LinearRegression
```

```python
vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1, 2),
    min_df=5
)
overview_text = vectorizer.fit_transform(train['overview'].fillna(''))
linreg = LinearRegression()
linreg.fit(overview_text, train['log_revenue'])
eli5.show_weights(linreg, vec=vectorizer, top=20, feature_filter=lambda x: x!= '<BIAS>')
```
